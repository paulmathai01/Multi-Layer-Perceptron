```python
# Importing required libraries
import pandas as pd             
import numpy as np                                    
import keras
from keras.models import Sequential                   
from keras.layers import Dense, Activation             
from keras.callbacks import History 
from sklearn.metrics import confusion_matrix          
from sklearn import preprocessing
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import LabelBinarizer      
from sklearn.model_selection import KFold
import seaborn as sns                                
import matplotlib.pyplot as plt                      
%matplotlib inline
np.random.seed(1)  
```

```python
# Load dataset from Keras
#KEEP DATA IN THE SAME DIRECTORY ***RELATIVE FILE PATH***
data = np.loadtxt("ion_data.txt")
(data_x, data_y) =(data[:, 0:7], data[:, 7:11])
```

```python
# Normalize x
data_x = data_x.astype(float) / np.amax(data_x)
```

```python
#Data Definitions
data_df = pd.DataFrame(data=data, columns=["Run","Pressure", "RF_Power","Electrode Gap","CCl4","He","O2","Etch Rate - Rp Å/min","Etch Uniformity – U (in %)","Oxide Selectivity - Sox","Photoresist Selectivity - Sph"])
data_x_df = pd.DataFrame(data=data_x, columns=["Run","Pressure", "RF_Power","Electrode Gap","CCl4","He","O2"])
data_y_df =  pd.DataFrame(data=data_y, columns=["Etch Rate - Rp Å/min","Etch Uniformity – U (in %)","Oxide Selectivity - Sox","Photoresist Selectivity - Sph"])

# Printing dimensions
print("Shape of X_train", data_df.shape)
print("Shape of y_train", data_x_df.shape)
print("Shape of X_test", data_y_df.shape)
```

```python
# General statistics for the data
data_df.describe()
```

```python
# Checking if data is correct
data_df.head()
```

```python
# Getting correlation plot of normalized dataset
corr = data_df.corr()
plt.figure(figsize = (10,10))
ax = sns.heatmap(corr, cmap=sns.diverging_palette(20, 220, n=200), annot = True, square=True)
```

```python
#The above correlation shows that Oxide Selectivity and Photoresit sensitivity are highly correlated. 
#Similarly RF Power,Etch Uniformity and Etch Rate,He Flow are correlated respectively.

#Normalising both X and Y to avoid variation in weightage for each 
normalized_X=(data_x_df-data_x_df.mean())/data_x_df.std()
normalized_y=(data_y_df-data_y_df.mean())/data_y_df.std()
normalized_X.head()
```

```python
# Box plot to explolre the distributiom
normalized_X.boxplot()
plt.xticks(rotation = 90)
```

```python
# Checking for Normalisatin of data
plt.figure(figsize = (8,8))
for i in normalized_y:
    sns.distplot(normalized_y[i], hist = False, label = i)
```

```python
X = np.asarray(normalized_X)
y = np.asarray(normalized_y)
print("Shape of input data: ", X.shape)
print("Shape of output data: ", y.shape)
```

```python
i_nodes = 100
h_nodes_1 = 200
```

```python
# Setting shuffle to True and making 42 K-Folds
kfold = KFold(42,True,1000)
# Training model on K-Folds

cvscores = []
kfold_num = 0
for train, test in kfold.split(X, y):
    # create model
    print("Fold %d " % kfold_num)
    model = Sequential()
    model.add(Dense(i_nodes, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(h_nodes_1, activation='relu'))
    model.add(Dense(y.shape[1], activation='linear'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error'])
    kfold_num += 1 
    # Fit the model
    history = model.fit(X[train], y[train], epochs=70, batch_size=1, verbose=2)
    # evaluate the model
    scores = model.evaluate(X[test], y[test], batch_size=1, verbose=0)
    print("Test %s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```

```python
true_var_whole_df = np.var(y, axis = 0)
y_predicted_whole_df = model.predict(X)
predicted_var_whole_df = np.var(y_predicted_whole_df, axis = 0)
print("Actual Variance: ", np.amax(true_var_whole_df))
print("Predicted Variance: ", np.amax(predicted_var_whole_df))      
print("Max Variance Predicted < Max variance Actual = ", 
     (np.amax(predicted_var_whole_df)) < (np.amax(true_var_whole_df)))
```

```python
results_var_whole_data = (predicted_var_whole_df) < (true_var_whole_df)
results_var_whole_data
print("Predicted Variance > True Variance =", False in results_var_whole_data)
```

```python
# Prepare the center points data for model evaluation
inputs_df = pd.DataFrame(normalized_X)
outputs_df = pd.DataFrame(normalized_y)
complete_df = pd.concat((inputs_df,outputs_df), axis = 1)
center_points_data = complete_df.iloc[[5, 22, 39, 42, 43, 45, 49, 50], :].values
center_points_data = pd.DataFrame(center_points_data)
X_center_points_df = center_points_data.iloc[:, 0:7].values
X_center_points = np.asarray(X_center_points_df)
y_center_points_df = center_points_data.iloc[:, 7:].values
y_center_points = np.asarray(y_center_points_df)
scores = model.evaluate(X_center_points, y_center_points, batch_size=1, verbose=0) 
print("Test %s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
# Defining the true center points
true_var_center_points = np.var(y_center_points, axis = 0)
# Predicting center points 
y_predicted_center_points = model.predict(X_center_points)
# Getting the variance for predicted center points
predicted_var_center_points = np.var(y_predicted_center_points, axis = 0)
# Comparison of predicted to actual max variance
print("Actual Variance at center points data: ", round((np.amax(true_var_center_points))*100,2),"%")
print("Predicted Variance at center points data: ", round((np.amax(predicted_var_center_points))*100,2),"%")      
# Check Ccndition for the conclusion 
print("Max Variance Predicted < Max variance Actual =  ", 
     (np.amax(predicted_var_center_points)) < (np.amax(true_var_center_points)))
```
